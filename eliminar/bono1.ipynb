### Descripción matemática
Las permutaciones son una forma de conteo para averiguar todas las formas en las que un conjunto de $n$ elementos puede reordenarse. La fórmula para lograr esto es:

$$
P(n) = n!
$$

> Esto solo aplica a elementos completamente diferentes y en fila. Si es en círculo usamos:
> 
> $$
> P_c(n) = (n - 1)!
> $$
> 
> Y para repetición usamos permutación con repetición:
>
> $$
> P_r(n) = \frac{n!}{n_1! n_2! n_3! \dots}
> $$
>
> Donde $n_1$, $n_2!$, etc. son las repeticiones de cada valor.

Por otro lado la k-permutación es un tipo de permutación en el que se seleccionan $k$ elementos de un conjunto de $n$ y plantea contar cuantas posibilidades hay. La fórmula para esto es:

$$
nPk = \frac{n!}{(n - k)!}
$$

Por definición $k \leq n$, porque si no, $n - k$ es negativo y por ende $(n - k)!$ no existiría.

En la permutación el orden importa siempre, por lo que elegir $(a,\ c,\ b)$, por ejemplo, es diferente a elegir $(b,\ c,\ a)$. La permutación normal puede considerarse como un caso especial de la k-permutación en la que $k = n$.

### Implementación de la permutación

#### 1. Permutaciones (factorial)
En la forma más básica, la librería `math` de Python ofrece la función `math.factorial` que podemos usar directamente

```python
import math as m

print(m.factorial(5))  # -> 120
```

Sin embargo vamos a crear dos alternativas, una iterativa y una recursiva. Ambas bien conocidas en programación. Usaremos anotaciones de tipo para mejor legibilidad y errores para manejo de forma externa.

```python
## Calcula el factorial de un entero dado usando iteración
def iterative_factorial(n: int) -> int:
    # Error de valor si es menor que cero
    if n < 0:
        raise ValueError("Valor no válido")

    fact: int = 1
    for i in range(n):
        fact *= i + 1

    return fact


## Calcula el factorial de un entero dado usando recursión
def recursive_factorial(n: int) -> int:
    # Error de valor si es menor que cero
    if n < 0:
        raise ValueError("Valor no válido")

    # Caso base
    if n == 0:
        return 1

    return n * recursive_factorial(n - 1)


# Prueba rápida
n = 5
print(f"Factorial iterativo de {n}: {iterative_factorial(n)}")
print(f"Factorial recursivo de {n}: {recursive_factorial(n)}")
```

Este último aprovecha una propiedad de los factoriales:

$$
n! = n(n - 1)!
$$

Si se repite este proceso con el factorial resutante de la derecha, podemos "desenrollar" un factorial hasta llegar a la forma base $0! = 1$, para luego proceder con los productos. Así funciona la versión recursiva.

La forma iterativa simplemente hace:

$$
n! = 1 \cdot 2 \cdot 3 \cdot \dots \cdot n
$$

#### 2. k-permutaciones
Para las k-permutaciones podríamos usar los factoriales anteriores, o el de la librería `math`. Sin embargo vamos a indagar de que van las k-permutaciones.

La fórmula es.

$$
nPk = \frac{n!}{(n - k)!}
$$

Si desenrollamos el factorial $n!$ hasta llegar a $(n - k)!$ obtenemos:

$$
\begin{align}
    nPk &= \frac{n \cdot (n - 1) \cdot (n - 2) \cdot \dots \cdot (n - k + 1) \cdot \cancel{(n - k)!}}{\cancel{(n - k)!}} \\
    nPk &= n \cdot (n - 1) \cdot (n - 2) \cdot \dots \cdot (n - k + 1) \\
\end{align}
$$

Esta es de hecho la forma original de la k-permutación, que es una aplicación del principio del producto para la elección sucesiva de opciones a la hora de combinar los elementos. Con esta fórmula podemos obtener la k-permutación usando un único bucle de forma más eficiente:

```python
## Implementa la k-permutación de un conjunto de n elementos para elegir k elementos
def k_permutation(n: int, k: int) -> int:
    # Error si alguno es menor que cero
    if k < 0 or n < 0:
        raise ValueError(f"Valores no válidos: {n}, {k}")

    # Error si k es mayor que n
    if k > n:
        raise ValueError(f"Valor de k no válido: {k}")

    # ? n+1 para abarcar n. range no es inclusivo al final
    result: int = 1
    for i in range(n - k + 1, n + 1):
        result *= i

    return result


n: int = 4
k: int = 3
print(
    f"Puedes seleccionar {k} elementos de un conjunto de {n} de {k_permutation(n, k)} formas diferentes"
)  # -> 12
```

### Aplicación de consola
Para la aplicación de consola usaremos prints e inputs de forma sencilla.

```python
print("+------------------------------------+")
print("|    Calculadora de permutaciones    |")
print("+------------------------------------+")
print("")


## Recibe el número del usuario con validaciones
def get_integer() -> int:
    value: int
    first_attemp: bool = True

    while True:
        i = input(
            f"{'Ingresa el numero de elementos' if first_attemp else 'Intenta de nuevo'}\n"
        )

        # Entrada cancelada
        if not i:
            print("Entrada cancelada")
            break

        # Intenta pasar a entero
        try:
            value = int(i)
            break
        except ValueError:
            print("[Error] Entrada no válida")
            first_attemp = False

    return value


n: int = get_integer()
fact1: int = None
fact2: int = None

# Intenta hacer el factorial
try:
    fact1 = iterative_factorial(n)
    fact2 = recursive_factorial(n)
    print(f"Factorial iterativo para {n}: {fact1}")
    print(f"Factorial recursivo para {n}: {fact2}")
except ValueError:
    print(f"[Error] Valor no válido para el factorial: {n}")
```
